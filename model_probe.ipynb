```python
from model.gnn_model import GraphClassifier 
from pipeline.graph_pipeline import get_train_test_loaders
import torch 

from IPython.display import SVG

from utils.molecule_visuzlizer import visualize_2d_quantitative_svg
from pipeline.graph_explainer import ModelExplainer
```

```python
train_loader, validation_loader = get_train_test_loaders(batch_size = 1024) 
```

```python
train_sample = next(iter(train_loader)) 
```

```python
classifier = GraphClassifier(
    node_attributes_shape= train_sample.x.shape[-1],
    edge_attributes_shape= train_sample.edge_attr.shape[-1],
    hidden_dim = 64
)

classifier.load_state_dict(torch.load("./trained_model/trained_model.pth"))
```

```python
index = 500

explainer = ModelExplainer(classifier) 

validation_batch = next(iter(validation_loader)) 

single_example = validation_batch.get_example(index)

node_scores, edge_scores = explainer.explain_graph(single_example)

smiles = single_example.smiles 
```

```python
classifier.predict(validation_batch.x, validation_batch.edge_index, validation_batch.edge_attr, validation_batch.batch)[index]
```
